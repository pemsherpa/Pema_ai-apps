```python
import pandas as pd
from IPython.display import display
import json
```

### Basic data reformatting

```python
df_2020 = pd.read_csv('../external_datasets/top_250_emissions/original/Public profit and emission database - 2020.csv')

cols_2020 = ["Company Name", "Industry", "Country/Territory ", "Scope 1 GHG emissions tons CO₂e", 
 "Scope 2 emissions  tons CO₂e", "Scope 3 emissions  tons CO₂e", 
 ' Sales/Revenue', 'Profit', 'Datasource emission data (name, link + page nr)']

cols_mapping_2020 = {"Company Name": "company_name", 
                     "Industry": "industry", 
                     "Country/Territory ": "country", 
                     "Scope 1 GHG emissions tons CO₂e": "scope_1", 
                     "Scope 2 emissions  tons CO₂e": "scope_2", 
                     "Scope 3 emissions  tons CO₂e": "scope_3", 
                     ' Sales/Revenue': 'sales_revenue', 
                     'Profit': 'profit', 
                     'Datasource emission data (name, link + page nr)': 'data_source'}

df_2020_clean = df_2020[["Company Name", "Industry", "Country/Territory ", "Scope 1 GHG emissions tons CO₂e", 
 "Scope 2 emissions  tons CO₂e", "Scope 3 emissions  tons CO₂e", 
 ' Sales/Revenue', 'Profit', 'Datasource emission data (name, link + page nr)']]

df_2020_clean = df_2020_clean.rename(columns=cols_mapping_2020)
df_2020_clean['year'] = 2020
df_2020_clean.head()
```

```python
df_2021 = pd.read_csv('../external_datasets/top_250_emissions/original/Public profit and emission database - 2021.csv')
df_2021 = df_2021.reset_index(drop=True).loc[:215,:]

cols_2021 = ['Company Name','2021 Scope 1 emissions tons CO₂e', 
             '2021 Scope 2 emissions tons CO₂e','2021 Scope 3 emissions tons CO₂e',
             '2021 Sales / Revenue (Million USD)', '2021 Profit (Million USD)',
             '2021-2022 Datasource emission data']

cols_mapping_2021 = {
                     'Company Name': 'company_name',
                     '2021 Scope 1 emissions tons CO₂e': 'scope_1', 
                     '2021 Scope 2 emissions tons CO₂e': 'scope_2',
                     '2021 Scope 3 emissions tons CO₂e': 'scope_3',
                     '2021 Sales / Revenue (Million USD)': 'sales_revenue', 
                     '2021 Profit (Million USD)': 'profit',
                     '2021-2022 Datasource emission data': 'data_source'
                     }

df_2021_clean = df_2021[cols_2021]
df_2021_clean = df_2021_clean.rename(columns=cols_mapping_2021)
df_2021_clean['year'] = 2021
df_2021_clean.head()
```

```python
df_2022 = pd.read_csv('../external_datasets/top_250_emissions/original/Public profit and emission database - 2022.csv')

df_2022.columns

cols_2022 = ['Company Name','2022 Scope 1 emissions tons CO₂e', '2022 Scope 2 emissions tons CO₂e',
       '2022 Scope 3 emissions tons CO₂e','2022 Sales / Revenue (Millions USD)', '2022 Profit (Millions USD)',
       '2021-2022 Datasource emission data']

cols_mapping_2022 = {'Company Name': 'company_name',
                     '2022 Scope 1 emissions tons CO₂e': 'scope_1', 
                     '2022 Scope 2 emissions tons CO₂e': 'scope_2',
                     '2022 Scope 3 emissions tons CO₂e': 'scope_3',
                     '2022 Sales / Revenue (Millions USD)': 'sales_revenue', 
                     '2022 Profit (Millions USD)': 'profit',
                     '2021-2022 Datasource emission data': 'data_source'}

df_2022_clean = df_2022[cols_2022]
df_2022_clean = df_2022_clean.rename(columns=cols_mapping_2022)
df_2022_clean['year'] = 2022
df_2022_clean.head()
```

### Concatenating the three data sets together

```python
df_concat = pd.concat([df_2020_clean, df_2021_clean, df_2022_clean], axis=0)
df_concat
```

### Cleaning up column names

```python
def clean_col_names(name: str) -> str:
    name = name.strip(' ')
    return name

df_concat['company_name'] = df_concat['company_name'].apply(clean_col_names)
```

```python
value_count = df_concat["company_name"].value_counts()
value_count
```

```python
with pd.option_context('display.max_rows', 500, 'display.max_columns', 10):
    display(value_count)
```

```python


# Sort DataFrame so that rows with non-NaN 'data_source' come first
df_sorted = df_concat.sort_values(by='data_source', na_position='last')

# Drop duplicates based on 'col1' and 'col2', keeping the first occurrence (which will have non-NaN 'data_source')
df_deduplicated = df_sorted.drop_duplicates(subset=['company_name', 'year'], keep='first')

# Reset the index for a cleaner DataFrame
df_deduplicated.reset_index(drop=True, inplace=True)

df_deduplicated
```
