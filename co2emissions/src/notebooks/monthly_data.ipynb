## Attempting to break up annual data into monthly data

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Using a basic sinusoidal model to model seasonal trends

```python
# Constants
months = np.arange(1, 13)
month_labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Given indices representing relative monthly emissions
indices = np.array([0.8, 0.8, 0.9, 1.2, 1.2, 1.2, 1.3, 1.3, 1.1, 1.0, 0.9, 0.8])

# Estimating parameters based on the provided indices
amplitude = 0.325
offset = 1.025
phase_shift = -np.pi/2  # Starts the sinusoidal wave at a low point

# Sinusoidal model
model_values = amplitude * np.sin(2 * np.pi * (months - 1) / 12 + phase_shift) + offset

# Plotting both the actual indices and the modeled values
plt.figure(figsize=(6, 4))
plt.plot(months, indices, marker='o', linestyle='-', color='blue', label='Actual Indices')
plt.plot(months, model_values, marker='x', linestyle='--', color='red', label='Sinusoidal Fit')
plt.title('Fit of Sinusoidal Model to Seasonal Emission Factors')
plt.xlabel('Month')
plt.ylabel('Relative Emission Factor')
plt.xticks(ticks=months, labels=month_labels)
plt.legend()
plt.show()
```
